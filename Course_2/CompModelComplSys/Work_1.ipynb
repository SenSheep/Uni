## Практика № 1 «Анализ движения точек» | Практика

В колесе радиуса R катится колесо радиуса r. Какую траекторию
описывает точка, расположенная на ободе колеса r?



Для получения траектории движения колеса, необходимо изменять
значение φ от 0 до 30, с шагом 0.01, вычислять координаты и представлять их
на графике. 

Задание 1:
Написать программу на python, реализующую расчет координат
соответствующих точек.
Задание 2:
Построить график полученной фигуры.

```python
import math as m
import matplotlib.pyplot as plt

x_values = []
y_values = []

R = 23
r = 5

step_size = 0.01
angle = 0

while angle <= 30 * m.pi:
    x = (R - r) * m.cos(angle) + r * m.cos(angle - (r/R) * angle)
    y = (R - r) * m.sin(angle) - r * m.sin(angle - (r/R) * angle)
    x_values.append(x)
    y_values.append(y)
    print(x, y)
    angle += step_size

plt.figure(figsize=(8, 8))
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Практика № 1 «Анализ движения точек» | Самостоятельная

Задание 1:

соответствующих точек с шагом 0.1 7.
Задание 2:
Построить график полученной фигуры.

```python
import numpy as np

B = 0.5

# Интервал и шаг
start = 0
end = 9 * np.pi
step = 0.1

# Генерация значений x и расчет координат y
x_values = np.arange(start, end + step, step)
y_values = np.sin(B * x_values)

# Вывод координат
for x, y in zip(x_values, y_values):
    print(f"x: {x:.2f}, y: {y:.2f}")

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
