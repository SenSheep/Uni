## Практика № 4 «ER-моделирование»
Построить ER-модели для следующих баз данных:
### 1. Социальная сеть:  
Предложите ER-модель для социальной сети, подобной Facebook или
Instagram. Определите сущности, такие как "Пользователь", "Пост", "Лайк",
"Комментарий" и другие подходящие сущности. Укажите атрибуты и
взаимосвязи между сущностями, например, какие посты опубликованы
каждым пользователем, кто оставил лайк к определенному посту и т.д.  


### 2. Библиотечная система:  
Разработайте ER-модель для системы управления библиотекой. Создайте
сущности, такие как "Книга", "Читатель", "Автор", "Выдача книги" и другие
необходимые сущности. Определите свойства каждой сущности и связи
между ними, например, какие книги находятся в наличии, какие книги
читатель взял на время и т.д.
