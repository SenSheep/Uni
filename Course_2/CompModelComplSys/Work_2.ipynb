## Практика № 2 «Модель движения фигуры» | Практика

Построить множество точек, соответствующих следующим фигурам:
Фигуры Ласажу:
Задание 1:
Написать программу на python, реализующую расчет координат
соответствующих точек.
Задание 2:
Построить график полученной фигуры.

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры фигуры Лассажа
A = 1
B = 1
a = 3
b = 5

# Интервал для параметра t
t_start = 0
t_end = 2 * np.pi
step = 0.01

# Генерация значений t и расчет координат x и y
t_values = np.arange(t_start, t_end + step, step)
x_values = A * np.sin(a * t_values)
y_values = B * np.cos(a * t_values)

# Вывод координат
for t, x, y in zip(t_values, x_values, y_values):
    print(f"t: {t:.2f}, x: {x:.2f}, y: {y:.2f}")

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

## Практика № 2 «Модель движения фигуры» | Самостоятельная

Построить множество точек, соответствующих следующим фигурам:
Гипотрохойда:
где φ изменяется от 0 до 2π.
Для получения траектории движения колеса, необходимо изменять
значение φ от 0 до 30, с шагом 0.01 вычислять координаты и представлять их
на графике.



Задание 1:
Написать программу на python, реализующую расчет координат
соответствующих точек.
Задание 2:
Построить график полученной фигуры.

```python
import math as m
import matplotlib.pyplot as plt

x_values = []
y_values = []

R = 23
r = 5

step_size = 0.01
angle = 0

while angle <= 30 * m.pi:
    x = (R - r) * m.cos(angle) + r * m.cos((R - r) / r * angle)
    y = (R - r) * m.sin(angle) - r * m.sin((R - r) / r * angle)
    x_values.append(x)
    y_values.append(y)
    print(x, y)
    angle += step_size

plt.figure(figsize=(8, 8))
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
