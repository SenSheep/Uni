(схема единственного деления)   


```python
import numpy as np

def gauss_elimination(A, b):
    n = len(b)
    # Создаем расширенную матрицу
    augmented_matrix = np.hstack([A, b.reshape(-1, 1)])

    # Прямой ход: приведение к верхнетреугольному виду
    for i in range(n):
        # Выбор ведущего элемента для предотвращения деления на ноль
        max_row = np.argmax(np.abs(augmented_matrix[i:, i])) + i
        augmented_matrix[[i, max_row]] = augmented_matrix[[max_row, i]]
        
        # Нормализация строки по ведущему элементу
        pivot = augmented_matrix[i, i]
        augmented_matrix[i] = augmented_matrix[i] / pivot
        
        # Обнуление элементов ниже ведущего
        for j in range(i + 1, n):
            factor = augmented_matrix[j, i]
            augmented_matrix[j] = augmented_matrix[j] - factor * augmented_matrix[i]

    # Обратный ход: нахождение переменных
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = augmented_matrix[i, -1] - np.sum(augmented_matrix[i, i+1:n] * x[i+1:n])
    
    return x

A = np.array([[2, 1, 4],
              [3, 2, 1],
              [1, 3, 3]])
b = np.array([16, 10, 16])

solution = gauss_elimination(A, b)
print("Решение системы:", solution)
```
