## Задача 1*.
Условие
Создайте csv-файл «rows_300.csv» со столбцами:
1) № - номер по порядку (от 1 до 300);
2) Секунда – текущая секунда на вашем ПК;
3) Микросекунда – текущая миллисекунда на часах.
На каждой итерации цикла искусственно приостанавливайте скрипт на 0,01 секунды.
Напоминание: Для работы с файлами подобного текстового формата потребуется
встроенная в Python библиотека csv.

```python
import csv
import time

# Определение функции для получения текущей секунды и микросекунды
def get_time():
    current_time = time.time()
    seconds = int(current_time)
    microseconds = int((current_time - seconds) * 1e6)
    return seconds, microseconds

# Создание CSV-файла
csv_file_path = 'rows_300.csv'

# Запись заголовков
headers = ['№', 'Секунда', 'Микросекунда']

with open(csv_file_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Запись заголовков
    writer.writerow(headers)
    
    # Запись данных
    for i in range(1, 301):
        seconds, microseconds = get_time()
        row = [i, seconds, microseconds]
        writer.writerow(row)
        
        # Искусственная пауза 0,01 секунды
        time.sleep(0.01)
```

## Задача 2*.
Условие
Установите библиотеку Pillow и при ее помощи в директории circles, созданной во
время выполнения функции, нарисуйте и сохраните 50 кругов радиусом 300 пикселей
случайных цветов в формате jpg на белом фоне (каждый круг - отдельный файл). Для
этого напишите функцию circles_generator (num_of_circles=50).
В первую очередь требуется установка модуля Pillow: $ pipinstallPillow
Случайным образом генерируйте цвета в палитре RGB и воспользуйтесь
классами Image, ImageDraw из установленной библиотеки. Чтобы нарисовать круг, нужно
применить метод ellipse() и задать координаты точек, соответствующие квадрату.
Необходимо вывести результат и сделать повторный вызов программы. Ответить
на вопрос: почему не будет ошибки при повторном вызове программы?

```python
from PIL import Image, ImageDraw
import os
import random


def circles_generator(num_of_circles=50):
    # Создание директории, если её нет
    directory = 'circles'
    if not os.path.exists(directory):
        os.makedirs(directory)

    for i in range(num_of_circles):
        # Генерация случайных цветов в формате RGB
        color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

        # Создание нового изображения с белым фоном
        image = Image.new('RGB', (600, 600), 'white')

        # Создание объекта ImageDraw для рисования
        draw = ImageDraw.Draw(image)
        
        # Координаты круга
        x = 0
        y = 0

        # Рисование круга радиусом 300 пикселей
        draw.ellipse([x, y, x + 600, y + 600], fill=color)

        # Сохранение изображения в файл
        file_path = os.path.join(directory, f'circle_{i + 1}.jpg')
        image.save(file_path)

        print(f'Создан файл: {file_path}')


# Вызываем функцию для создания 50 кругов
circles_generator(50)
```
