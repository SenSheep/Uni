## –ó–∞–¥–∞—á–∞ 1
–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å "–§–∏–ª—å–º" —Å–æ —Å–≤–æ–π—Å—Ç–≤–∞–º–∏ "–Ω–∞–∑–≤–∞–Ω–∏–µ", "–≥–æ–¥ –≤—ã–ø—É—Å–∫–∞" –∏ "—Ä–µ–π—Ç–∏–Ω–≥". –°–æ–∑–¥–∞–π—Ç–µ
—Å–ø–∏—Å–æ–∫ —Ñ–∏–ª—å–º–æ–≤ –∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –µ–≥–æ —Å–Ω–∞—á–∞–ª–∞ –ø–æ –≥–æ–¥—É –≤—ã–ø—É—Å–∫–∞, –∞ –∑–∞—Ç–µ–º –ø–æ —Ä–µ–π—Ç–∏–Ω–≥—É,
–∏—Å–ø–æ–ª—å–∑—É—è –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü–∏–∏ –≤ –º–µ—Ç–æ–¥–µ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.

```python
class Film:
    def __init__(self, title, year, rating):
        self.title = title
        self.year = year
        self.rating = rating

# –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ —Ñ–∏–ª—å–º–æ–≤
films = [
    Film("–§–∏–ª—å–º1", 2000, 8.5),
    Film("–§–∏–ª—å–º2", 1995, 7.8),
    Film("–§–∏–ª—å–º3", 2010, 9.2),
    # –î–æ–±–∞–≤—å—Ç–µ —Å–≤–æ–∏ —Ñ–∏–ª—å–º—ã –ø–æ –∞–Ω–∞–ª–æ–≥–∏–∏
]

# –°–æ—Ä—Ç–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ —Å–Ω–∞—á–∞–ª–∞ –ø–æ –≥–æ–¥—É –≤—ã–ø—É—Å–∫–∞, –∞ –∑–∞—Ç–µ–º –ø–æ —Ä–µ–π—Ç–∏–Ω–≥—É
films.sort(key=lambda x: x.year)
films.sort(key=lambda x: x.rating, reverse=True)

# –í—ã–≤–æ–¥–∏–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫
for film in films:
    print(f"{film.title} ({film.year}) - –†–µ–π—Ç–∏–Ω–≥: {film.rating}")
```

## –ó–∞–¥–∞—á–∞ 2
–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å "–¢–æ–≤–∞—Ä" —Å–æ —Å–≤–æ–π—Å—Ç–≤–∞–º–∏ "–Ω–∞–∑–≤–∞–Ω–∏–µ" –∏ "—Ü–µ–Ω–∞". –°–æ–∑–¥–∞–π—Ç–µ —Å–ø–∏—Å–æ–∫ —Ç–æ–≤–∞—Ä–æ–≤ –∏
–∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü–∏—é –≤ —Ñ—É–Ω–∫—Ü–∏—è—Ö `map()` –∏ `reduce()` –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Å—É–º–º–∞—Ä–Ω–æ–π
—Å—Ç–æ–∏–º–æ—Å—Ç–∏ –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤.

```python
from functools import reduce

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

# –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ —Ç–æ–≤–∞—Ä–æ–≤
products = [
    Product("–¢–æ–≤–∞—Ä1", 100),
    Product("–¢–æ–≤–∞—Ä2", 150),
    Product("–¢–æ–≤–∞—Ä3", 80),
    # –î–æ–±–∞–≤—å—Ç–µ —Å–≤–æ–∏ —Ç–æ–≤–∞—Ä—ã –ø–æ –∞–Ω–∞–ª–æ–≥–∏–∏
]

# –ò—Å–ø–æ–ª—å–∑—É–µ–º map() –∏ lambda-—Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ —Ü–µ–Ω —Ç–æ–≤–∞—Ä–æ–≤
prices = list(map(lambda x: x.price, products))

# –ò—Å–ø–æ–ª—å–∑—É–µ–º reduce() –∏ lambda-—Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Å—É–º–º–∞—Ä–Ω–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤
total_price = reduce(lambda x, y: x + y, prices, 0)

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(f"–°—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤: {total_price}")
```

```python
def genalgho(gen): # –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ–≥–æ –ø–æ–∫–æ–ª–µ–Ω–∏—è
    ngen = []
    z = bgen = 0
    for i in gen:
        zt = (i[0] - 2 * i[1])/(i[0] ** 2 + i[1] ** 2 + 1)
        z = max(z, zt)
        if z == zt: # –õ—É—á—à–∏–π –≥–µ–Ω
            bgen = i

    for i in range((len(gen))): # –ü–µ—Ä–µ—Ç–∞—Å–æ–≤–∫–∞ —Ö—Ä–æ–º–æ—Å–æ–º
        if bgen == gen[i]:
            continue
        else:
            ngen.append([bgen[0], gen[i][1]])
    for i in range((len(gen))):
        if bgen == gen[i]:
            continue
        else:
            ngen.append([bgen[1], gen[i][0]])

    return(ngen)

def bestgen(gen): # –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –ª—É—á—à–∏—Ö –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–µ–ª–µ–π –ø–æ–∫–æ–ª–µ–Ω–∏–π
    ngen = []
    for h in range(3):
        z = bgen = 0
        for i in gen:
            zt = (i[0] - 2 * i[1])/(i[0] ** 2 + i[1] ** 2 + 1)
            z = max(z, zt)
            if z == zt:
                bgen = i

        for i in range((len(gen))):
            if bgen == gen[i]:
                ngen.append(gen[i])
                break
        gen.remove(bgen)
    
    return(ngen)

def bZaZ(gen): # –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ Z –∏ –ª—É—á—à–µ–≥–æ Z
    s = []
    z = 0
    for i in gen:
        zt = (i[0] - 2 * i[1])/(i[0] ** 2 + i[1] ** 2 + 1)
        s.append(zt)
        z = max(z, zt)
    zt = sum(s)/len(s)
    
    return(round(z, 2), round(zt, 2))

gen = [[-1, 0], [0, -1], [2, -2], [3, 1]] # –ü—Ä–∏–º–µ—Ä –ø–æ–∫–æ–ª–µ–Ω–∏—è
for x in range(3): # –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –≥–µ–Ω–æ–≤ –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ 3 –ø–æ–∫–æ–ª–µ–Ω–∏–π
    newgenraw = genalgho(gen)
    bgen = bestgen(newgenraw)
    bz, az = bZaZ(bgen)
    print(f"–ü–æ–∫–æ–ª–µ–Ω–∏–µ {x}: {bgen}; –°—Ä–µ–¥–Ω–µ–µ Z: {az}; –õ—É—á—à–µ–µ Z: {bz}")
    gen = bgen
```
