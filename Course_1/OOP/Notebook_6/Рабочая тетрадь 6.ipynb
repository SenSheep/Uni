## Задача 1.
На мебельном производстве ассортимент выпускаемой продукции имеет базовые
товары:
- столы, у которых много всяких полей, методов, есть своя внутренняя логика;
- шкафы, у которых тоже много полей, методов, есть своя внутренняя логика.

Есть также тумбочки, которые могут функционально использоваться как столы,
так и как шкафы и также имеют свою специфическую логику, поля, методы.
Попробуйте реализовать множественное наследование с помощью super и без него

```python
# С использованием super
class Table:
    def __init__(self, material, color, shape):
        self.material = material
        self.color = color
        self.shape = shape

    def describe(self):
        print(f"This table is made of {self.material}, has a {self.color} color, and has a {self.shape} shape.")

class Wardrobe:
    def __init__(self, material, color, num_doors):
        self.material = material
        self.color = color
        self.num_doors = num_doors

    def describe(self):
        print(f"This wardrobe is made of {self.material}, has a {self.color} color, and has {self.num_doors} doors.")

class Nightstand(Table, Wardrobe):
    def __init__(self, material, color, shape, num_doors, drawers):
        super().__init__(material, color, shape)
        Wardrobe.__init__(self, material, color, num_doors)
        self.drawers = drawers

    def describe(self):
        print(f"This nightstand is made of {self.material}, has a {self.color} color, "
              f"has a {self.shape} shape, {self.num_doors} doors, and {self.drawers} drawers.")


# Пример использования
nightstand = Nightstand("Wood", "Brown", "Square", 2, 2)
nightstand.describe()
```

```python
# Без использования super
class Table:
    def __init__(self, material, color, shape):
        self.material = material
        self.color = color
        self.shape = shape

    def describe(self):
        print(f"This table is made of {self.material}, has a {self.color} color, and has a {self.shape} shape.")

class Wardrobe:
    def __init__(self, material, color, num_doors):
        self.material = material
        self.color = color
        self.num_doors = num_doors

    def describe(self):
        print(f"This wardrobe is made of {self.material}, has a {self.color} color, and has {self.num_doors} doors.")

class Nightstand(Table, Wardrobe):
    def __init__(self, material, color, shape, num_doors, drawers):
        Table.__init__(self, material, color, shape)
        Wardrobe.__init__(self, material, color, num_doors)
        self.drawers = drawers

    def describe(self):
        print(f"This nightstand is made of {self.material}, has a {self.color} color, "
              f"has a {self.shape} shape, {self.num_doors} doors, and {self.drawers} drawers.")


# Пример использования
nightstand = Nightstand("Wood", "Brown", "Square", 2, 2)
nightstand.describe()
```

## Задача 2
Построить иерархию классов согласно схеме наследования, приведенной на рисунке
ниже, по любой предметной области. Каждый класс должен содержать необходимые
конструкторы и методы работы с полями классов. Функция main() должна иллюстрировать
работу с объектами всех созданных классов.


```python
class C:
    def __init__(self, field_c):
        self.field_c = field_c

    def method_c(self):
        print("Method in class C")


class D1(C):
    def __init__(self, field_d1, field_c):
        super().__init__(field_c)
        self.field_d1 = field_d1

    def method_d1(self):
        print("Method in class D1")


class D2(D1):
    def __init__(self, field_d2, field_d1, field_c):
        super(D2, self).__init__(field_d1, field_c)
        self.field_d2 = field_d2

    def method_d2(self):
        print("Method in class D2")


class D3(D2):
    def __init__(self, field_d3, field_d2, field_d1, field_c):
        super().__init__(field_d2, field_d1, field_c)
        self.field_d3 = field_d3

    def method_d3(self):
        print("Method in class D3")


class B1:
    def __init__(self, field_b1):
        self._field_b1 = field_b1  # _field_b1 - защищенное поле

    def method_b1(self):
        print("Method in class B1")


class B2(B1):
    def __init__(self, field_b2, field_b1):
        super().__init__(field_b1)
        self.__field_b2 = field_b2  # __field_b2 - приватное поле

    def method_b2(self):
        print("Method in class B2")


# Функция main для иллюстрации работы с объектами классов
def main():
    obj_d3 = D3("D3", "D2", "D1", "C")
    obj_d3.method_c()
    obj_d3.method_d1()
    obj_d3.method_d2()
    obj_d3.method_d3()

    obj_b2 = B2("B2", "B1")
    obj_b2.method_b1()
    obj_b2.method_b2()


# Вызываем функцию main
main()
```
